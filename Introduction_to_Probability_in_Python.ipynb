```python
import numpy as np
import matplotlib.pyplot as plt

# Introduction to Probability for Deep Learning using Python

# Probability theory is a fundamental concept in deep learning, enabling us to model uncertainty and make predictions. 
# In this code example, we'll explore the basics of probability using Python, focusing on the normal distribution and discrete distributions. 
# We'll generate random samples, visualize the distributions, calculate mean and standard deviation, and demonstrate working with discrete distributions.

# Generating random numbers from the normal distribution
num_samples = 1000
mean = 0
std = 1
samples = np.random.normal(mean, std, num_samples)

# Plotting the histogram of the samples
plt.figure(figsize=(8, 4))
plt.hist(samples, bins=30, density=True, alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Normal Distribution')
plt.grid(True)
plt.show()

# Calculating the mean and standard deviation
sample_mean = np.mean(samples)
sample_std = np.std(samples)

# Printing the calculated mean and standard deviation
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)

# Generating random numbers from a discrete distribution
probabilities = [0.1, 0.3, 0.2, 0.4]
values = [1, 2, 3, 4]
discrete_samples = np.random.choice(values, num_samples, p=probabilities)

# Plotting the histogram of the discrete samples
unique_values, value_counts = np.unique(discrete_samples, return_counts=True)
probability_counts = value_counts / num_samples

plt.figure(figsize=(8, 4))
plt.bar(unique_values, probability_counts)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Distribution')
plt.grid(True)
plt.show()

# Calculating the probability of a specific value
target_value = 3
target_probability = probability_counts[np.where(unique_values == target_value)]
print("Probability of value", target_value, ":", target_probability[0])
```
