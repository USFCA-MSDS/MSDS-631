```python
import torch
import matplotlib.pyplot as plt

# Create a tensor with requires_grad=True to enable auto differentiation
x = torch.tensor(2.0, requires_grad=True)

# Define a mathematical function
def f(x):
    return 3 * x**2 + 2 * x + 1

# Compute the function value
y = f(x)
print("Function value:", y.item())

# Compute the derivative using auto differentiation
y.backward()
x_grad = x.grad # the derivative of f with respect to x evaluated at 2 
# the grad is 6x + 2 and @ 2 this becomes 14.
print("Derivative value:", x_grad.item())

# Visualize the function and its tangent line
x_vals = torch.linspace(-5, 5, 100)
y_vals = f(x_vals)
tangent_line = x_grad.item() * (x_vals - x.item()) + y.item()

plt.plot(x_vals, y_vals, label='f(x)')
plt.plot(x_vals, tangent_line, label='Tangent line')
plt.scatter(x.item(), y.item(), color='red', label='Point (x, f(x))')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
