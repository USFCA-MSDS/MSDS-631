# Bijectors in Torch




In Normalizing Flows (NF), a **bijector** is a key component used to define the transformation between a simple probability distribution (e.g., a standard Gaussian distribution) and a more complex target distribution. The purpose of a bijector is to model a bijection (a one-to-one mapping) between the input space and the output space, allowing for flexible and expressive transformations.

In the context of NF in PyTorch, a bijector is typically implemented as a PyTorch module that inherits from torch.nn.Module. It provides methods for forward and inverse transformations, allowing data to be mapped between the input and output spaces.

The forward transformation maps samples from the simple distribution (e.g., the standard Gaussian) to samples from the target distribution. It takes an input sample and applies a series of parameterized transformations (often implemented as neural networks) to produce the transformed output sample.

The inverse transformation maps samples from the target distribution back to the simple distribution. It takes an output sample and applies the inverse of the forward transformations to reconstruct the original input sample.

The bijector is trained by optimizing the parameters of the transformations to maximize the likelihood of the observed data. This is typically done by minimizing a loss function that incorporates both the likelihood of the target distribution and a regularization term.

By composing multiple bijectors together, complex and flexible transformations can be achieved. Each bijector contributes to the overall transformation, and the composition allows for modeling a wide variety of target distributions.

The use of bijectors in NF provides a way to model complex data distributions while maintaining tractability for density estimation and sampling. They offer a powerful framework for generative modeling and enable various applications such as density estimation, data generation, and latent space modeling.





```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.distributions as distributions
import numpy as np
import matplotlib.pyplot as plt

# Define a custom Bijector class
class MyBijector(nn.Module):
    def __init__(self):
        super(MyBijector, self).__init__()

    def forward(self, x):
        # Transformation function
        y = torch.sin(x)
        return y

    def inverse(self, y):
        # Inverse transformation function
        x = torch.asin(y)
        return x

# Generate some random data
num_samples = 1000
x_samples = np.random.uniform(-1, 1, size=num_samples)

# Convert the data to tensors
x_tensor = torch.tensor(x_samples, dtype=torch.float32)

# Create an instance of the Bijector
bijector = MyBijector()

# Apply the forward transformation
y_tensor = bijector.forward(x_tensor)

# Apply the inverse transformation
x_reconstructed = bijector.inverse(y_tensor)

# Plot the original and reconstructed data
plt.figure(figsize=(8, 6))
plt.scatter(x_tensor, np.zeros_like(x_samples), color='blue', label='Original Data')
plt.scatter(x_reconstructed, np.zeros_like(x_samples), color='red', label='Reconstructed Data')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bijector Example')
plt.legend()
plt.show()
```





In the above example, we define a custom MyBijector class that inherits from the nn.Module class. The class includes a forward method that performs the forward transformation (y = sin(x)) and an inverse method that performs the inverse transformation (x = asin(y)).

We generate random data (x_samples) and convert it to a PyTorch tensor (x_tensor). We then create an instance of the MyBijector class.

We apply the forward transformation by passing the x_tensor through the forward method of the bijector, resulting in y_tensor. We then apply the inverse transformation by passing y_tensor through the inverse method of the bijector, reconstructing x_reconstructed.

Finally, we plot both the original data (x_tensor) and the reconstructed data (x_reconstructed) to visualize the effect of the bijective transformation.

Note that this example uses a simple transformation for illustration purposes. In practice, more complex bijectors can be designed and applied to learn sophisticated data transformations.
