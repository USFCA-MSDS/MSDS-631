# An introduction to the gradient descent algorithm

Gradient descent is an iterative optimization algorithm that updates the parameter values in the direction of steepest descent of the objective function. By repeatedly calculating the gradient and updating the parameters, it aims to find the optimal parameter values that minimize the objective function.

# Next we present an example in python that illuratet the general steps of the algorithm.

* We define the objective function objective(x), which represents the function we want to minimize. 

* We define the gradient function gradient(x), which represents the derivative of the objective function with respect to the parameter x.

* We set the initial guess x0, learning rate (learning_rate), and number of iterations (num_iterations) for the gradient descent algorithm.

* We perform gradient descent by iteratively updating the parameter x using the gradient and learning rate. In each iteration, we calculate the gradient of the objective function with respect to x and update x by subtracting the learning rate multiplied by the gradient.

* Finally, we print the optimized parameter value, which represents the result of the gradient descent optimization.




```python
import numpy as np
from scipy.optimize import minimize

# Define the objective function
def objective(x):
    return x[0]**2 + x[1]**2

# The exact minimum for the objective function objective(x) = x[0]**2 + x[1]**2 is at the point (0, 0).

# Define the gradient of the objective function
def gradient(x):
    return np.array([2 * x[0], 2 * x[1]])

# Perform gradient descent optimization
def gradient_descent(objective, gradient, x0, learning_rate, num_iterations):
    x = x0
    for i in range(num_iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
    return x

# Set the initial guess and hyperparameters for gradient descent
x0 = np.array([1.0, 1.0])
learning_rate = 0.1
num_iterations = 100

# Perform gradient descent optimization
gd_result = gradient_descent(objective, gradient, x0, learning_rate, num_iterations)

# Perform optimization using scipy.optimize
scipy_result = minimize(objective, x0)


print(" Exact Min for the objective function:", (0,0))
print("")

print("The optimization results and the computed min:")

# Print the optimization results
print("Gradient Descent Result:")
print("  Objective Value:", objective(gd_result))
print("  Optimized Parameters:", gd_result)
print()
print("Scipy Optimization Result:")
print("  Objective Value:", scipy_result.fun)
print("  Optimized Parameters:", scipy_result.x)
```

In this updated code:

*  We define the objective function objective(x) and its gradient function gradient(x). The objective function remains the same as before (x[0]**2 + x[1]**2), while the gradient is computed analytically (2 * x[0] and 2 * x[1]).

*  We define the gradient_descent function that performs gradient descent optimization. It takes the objective function, gradient function, initial guess, learning rate, and number of iterations as input. It iteratively updates the parameter values using the gradient and learning rate.

*  We set the initial guess x0, learning rate (learning_rate), and number of iterations (num_iterations) for gradient descent.

*  We call the gradient_descent function to perform the gradient descent optimization and store the result in gd_result.

* We use the minimize function from scipy.optimize to perform the optimization using the BFGS algorithm. The objective function and initial guess are provided as inputs.

The optimization results from both gradient descent (gd_result) and scipy.optimize (scipy_result) are printed, displaying the optimized objective value and the optimized parameters.

This updated code demonstrates the use of gradient descent optimization using NumPy and compares it with the optimization result obtained using the scipy.optimize module.

# Example 2 : visulize gradient decsent

```python
import numpy as np
import matplotlib.pyplot as plt

# Gradient Descent

# Define the objective function
def objective(x):
    return x**2  # Example objective function: x^2

# Define the derivative of the objective function
def gradient(x):
    return 2 * x  # Derivative of x^2 is 2x

# Set the initial guess and hyperparameters
x0 = 5.0  # Initial guess
learning_rate = 0.1  # Learning rate
num_iterations = 10  # Number of iterations

# Perform gradient descent
x = x0
x_vals = [x]
y_vals = [objective(x)]
for i in range(num_iterations):
    grad = gradient(x)
    x = x - learning_rate * grad
    x_vals.append(x)
    y_vals.append(objective(x))

# Create the plot
x_vals = np.array(x_vals)
y_vals = np.array(y_vals)

# Plot the optimization path
plt.plot(x_vals, y_vals, '-o', color='blue', label='Optimization Path')

# Plot the objective function
x_range = np.linspace(-6, 6, 100)  # Range of x values for plotting
y_range = objective(x_range)
plt.plot(x_range, y_range, color='red', label='Objective Function')

# Plot the initial guess
plt.plot(x0, objective(x0), 'o', color='green', label='Initial Guess')

# Plot the final optimized point
plt.plot(x, objective(x), 'o', color='purple', label='Final Optimized Point')

plt.xlabel('x')
plt.ylabel('Objective Function')
plt.title('Gradient Descent Optimization')
plt.grid(True)
plt.legend()
plt.show()
```

* The optimization path is plotted with blue lines and markers, representing the values of x and the objective function (y) at each iteration.

* The objective function is plotted in red, providing a visual representation of the curve.

* The initial guess is marked with a green dot, and the final optimized point is marked with a purple dot.



# 2d Example

```python
import numpy as np
import matplotlib.pyplot as plt

# Gradient Descent for 2D Function

# Define the objective function
def objective(x):
    return x[0]**2 + x[1]**2  # Example objective function: x1^2 + x2^2

# Define the gradient of the objective function
def gradient(x):
    return np.array([2 * x[0], 2 * x[1]])  # Gradient of x1^2 + x2^2 is [2x1, 2x2]

# Set the initial guess and hyperparameters
x0 = np.array([4.0, -3.0])  # Initial guess [x1, x2]
learning_rate = 0.1  # Learning rate
num_iterations = 10  # Number of iterations

# Perform gradient descent
x = x0
x_vals = [x]
y_vals = [objective(x)]
for i in range(num_iterations):
    grad = gradient(x)
    x = x - learning_rate * grad
    x_vals.append(x)
    y_vals.append(objective(x))

# Create a meshgrid of x1 and x2 values for contour plot
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = X1**2 + X2**2

# Create the contour plot
plt.figure(figsize=(8, 6))
plt.contour(X1, X2, Z, levels=20, cmap='jet')
plt.colorbar(label='Objective Function')
plt.plot(x_vals[0][0], x_vals[0][1], 'go', label='Initial Guess')
plt.plot(x_vals[-1][0], x_vals[-1][1], 'ro', label='Final Optimized Point')
plt.plot([x[0] for x in x_vals], [x[1] for x in x_vals], '-o', color='blue', label='Optimization Path')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()
```

In this updated example:

* The objective function is a simple quadratic function x1^2 + x2^2 in two variables x1 and x2.

*  The gradient of the objective function is defined accordingly.

* The initial guess is set to [4.0, -3.0], and the learning rate and number of iterations are specified.

* Gradient descent is performed, and the values of x and the objective function y are stored at each iteration.

* A meshgrid of x1 and x2 values is created to generate the contour plot.
 
* The contour plot is created using plt.contour, representing the objective function landscape.

* The initial guess and final optimized point are marked with green and red dots, respectively.

* The optimization path is plotted using blue lines and markers to illustrate the trajectory of gradient descent.
