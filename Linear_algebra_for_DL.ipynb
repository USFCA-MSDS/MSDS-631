Linear algebra plays a fundamental role in deep learning, enabling us to perform various operations on data to extract meaningful patterns and make accurate predictions. In this code example, we explore the application of linear transformations with a focus on non-linear activation using the Rectified Linear Unit (ReLU). By understanding and implementing these concepts, we can gain insights into how deep learning models learn and process data.

The Code:
The code begins by defining an input shape as a set of points in a two-dimensional space. These points represent the initial object that we aim to transform using linear operations. We then visualize the initial shape to provide a clear reference.

The first transformation applied is scaling, which involves multiplying each coordinate of the object by scaling factors. By specifying appropriate scaling factors, we can stretch or shrink the object along different dimensions. The scaled shape is plotted to visualize the transformation.

Next, we introduce rotation as a linear transformation. We rotate the original shape by a specified angle using a rotation matrix. The rotated shape is then displayed, showcasing the effect of rotation on the object.

Finally, we explore the concept of shearing, a linear transformation that skews the shape along different axes. By applying a shearing matrix to the initial shape, we observe how the object is distorted. The sheared shape is plotted for visualization.

To incorporate non-linearity into the transformations, we apply the ReLU activation function. The ReLU function sets negative values to zero while leaving positive values unchanged. We visualize both the transformed shapes before and after applying the ReLU activation, providing a clear understanding of how non-linearity impacts the final outcome.


```python
import numpy as np
import matplotlib.pyplot as plt

# Linear Algebra in Deep Learning: Linear Transformations

# Define the input shape
points = np.array([[1, 1],
                   [1, 3],
                   [3, 3],
                   [3, 1]])

# Plot the input shape
plt.figure(figsize=(5, 5))
plt.plot(points[:, 0], points[:, 1], '-o', color='b')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Input Shape')
plt.grid(True)
plt.show()


# Linear Transformation: Scaling

# Define the scaling factor
scale_x = 2
scale_y = 0.5

# Create the scaling matrix
scaling_matrix = np.array([[scale_x, 0],
                           [0, scale_y]])

# Apply the scaling transformation
scaled_points = np.dot(points, scaling_matrix)

# Plot the scaled shape
plt.figure(figsize=(5, 5))
plt.plot(scaled_points[:, 0], scaled_points[:, 1], '-o', color='r')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scaled Shape')
plt.grid(True)
plt.show()


# Linear Transformation: Rotation

# Define the rotation angle (in radians)
angle = np.pi / 4

# Create the rotation matrix
rotation_matrix = np.array([[np.cos(angle), -np.sin(angle)],
                            [np.sin(angle), np.cos(angle)]])

# Apply the rotation transformation
rotated_points = np.dot(points, rotation_matrix)

# Plot the rotated shape
plt.figure(figsize=(5, 5))
plt.plot(rotated_points[:, 0], rotated_points[:, 1], '-o', color='g')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Rotated Shape')
plt.grid(True)
plt.show()


# Linear Transformation: Shearing

# Define the shearing factors
shear_x = 1.5
shear_y = 0.5

# Create the shearing matrix
shearing_matrix = np.array([[1, shear_x],
                            [shear_y, 1]])

# Apply the shearing transformation
sheared_points = np.dot(points, shearing_matrix)

# Plot the sheared shape
plt.figure(figsize=(5, 5))
plt.plot(sheared_points[:, 0], sheared_points[:, 1], '-o', color='m')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sheared Shape')
plt.grid(True)
plt.show()
```

In this example:

* The input shape is defined as a set of points in 2D space.

* The input shape is plotted to visualize the original shape.

* Linear transformations such as scaling, rotation, and shearing are applied to the input shape using matrix multiplication.

* Each transformed shape is plotted individually to visualize how the linear transformation changes the geometry of the input shape.

* Detailed comments are provided throughout the code to explain the linear algebra concepts and the effect of each transformation.

*  By visualizing the input shape and the transformed shapes using matplotlib, you can observe how linear transformations such as scaling, rotation, and shearing alter the size, orientation, and shape of the original input.
